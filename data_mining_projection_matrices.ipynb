```python
MIN_DENSITY = 0.5
MIN_VOLUME = 0.1

filtered_unionizes = structure_unionizes[structure_unionizes.projection_density > MIN_DENSITY]
filtered_unionizes = filtered_unionizes[filtered_unionizes.volume > MIN_VOLUME]
```

## Projection Matrix

```python
def get_structure_name_from_id(s_id):
    s = structure_tree.get_structures_by_id([s_id])[0]
    return s['name']

# hemisphere_id	ID of the hemisphere (1 = left, 2 = right, 3 = both)
def select_highest_projections(p_matrix, number_to_select=10, hemisphere='both', change_labels=False):
    if hemisphere=='left': hemisphere = 1
    elif hemisphere=='right': hemisphere = 2
    elif hemisphere=='both': hemisphere = 3
    matrix = p_matrix['matrix']
    exp_ids = p_matrix['rows']
    area_info = p_matrix['columns']
    
    sum_along_exps = np.sum(matrix, axis=0)
    highest_inds = np.argsort(sum_along_exps)[-number_to_select:]
    highest_inds = highest_inds[::-1]
    
    out_matrix = {}
    out_matrix['matrix'] = matrix[:,highest_inds]
    out_matrix['columns'] = []
    out_matrix['rows'] = exp_ids
    for i in highest_inds: 
        out_matrix['columns'].append(area_info[i])
        if change_labels: out_matrix['columns'][-1]['label'] = get_structure_name_from_id(out_matrix['columns'][-1]['structure_id'])
    
    return out_matrix

def get_projection_target_areas_from_set(set_id):
    return list(pd.DataFrame(structure_tree.get_structures_by_set_id([set_id])).id)
```

```python
experiment_ids = [e['id'] for e in experiments]
# children_areas = structure_tree.child_ids([AREA_ID])[0]
# projection_areas = children_areas
# projection_areas = list(coarse_set_df.id)
projection_areas = get_projection_target_areas_from_set(3) # coarse
# projection_areas = get_projection_target_areas_from_set(167587189) # identical to connectivity.brain-map.org

proj_matrix = mcc.get_projection_matrix(experiment_ids = experiment_ids, 
                               projection_structure_ids = projection_areas,
                               parameter = 'projection_density')

# proj_matrix
```

```python
proj_matrix = select_highest_projections(proj_matrix, number_to_select=10, hemisphere='both', change_labels=False)
```

```python
row_labels = proj_matrix['rows'] # these are just experiment ids
column_labels = [ c['label'] for c in proj_matrix['columns'] ] 
matrix = proj_matrix['matrix']

fig, ax = plt.subplots(figsize=(15,15), facecolor='white')
heatmap = ax.pcolor(matrix, cmap=plt.cm.afmhot)

# put the major ticks at the middle of each cell
ax.set_xticks(np.arange(matrix.shape[1])+0.5, minor=False)
ax.set_yticks(np.arange(matrix.shape[0])+0.5, minor=False)

ax.set_xlim([0, matrix.shape[1]])
ax.set_ylim([0, matrix.shape[0]])          

# want a more natural, table-like display
ax.invert_yaxis()
# ax.xaxis.tick_top()

ax.set_xticklabels(column_labels, minor=False)
ax.set_yticklabels(row_labels, minor=False)

plt.xlabel('Areas')
plt.ylabel('Experiment IDs')
```

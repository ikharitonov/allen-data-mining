### Downloading projections data

```python
experiment_id = 294481346
# experiment_id = 159753308
```

```python
# projection density: number of projecting pixels / voxel volume
pd, pd_info = mcc.get_projection_density(experiment_id)

# injection density: number of projecting pixels in injection site / voxel volume
ind, ind_info = mcc.get_injection_density(experiment_id)

# injection fraction: number of pixels in injection site / voxel volume
inf, inf_info = mcc.get_injection_fraction(experiment_id)

# data mask:
# binary mask indicating which voxels contain valid data
dm, dm_info = mcc.get_data_mask(experiment_id)

template, template_info = mcc.get_template_volume()
annot, annot_info = mcc.get_annotation_volume()

# in addition to the annotation volume, you can get binary masks for individual structures
# in this case, we'll get one for the isocortex
cortex_mask, cm_info = mcc.get_structure_mask(315)

print(pd_info)
print(pd.shape, template.shape, annot.shape)
```

```python
# data mask - valid pixel data
plt.imshow(dm.max(axis=0), cmap='hot', aspect='equal')
```

```python
# compute the maximum intensity projection (along the anterior-posterior axis) of the projection data
pd_mip = pd.max(axis=0)
ind_mip = ind.max(axis=0)
inf_mip = inf.max(axis=0)

# show that slice of all volumes side-by-side
f, pr_axes = plt.subplots(1, 3, figsize=(15, 6))

pr_axes[0].imshow(pd_mip, cmap='hot', aspect='equal')
pr_axes[0].set_title("projection density MaxIP")

pr_axes[1].imshow(ind_mip, cmap='hot', aspect='equal')
pr_axes[1].set_title("injection density MaxIP")

pr_axes[2].imshow(inf_mip, cmap='hot', aspect='equal')
pr_axes[2].set_title("injection fraction MaxIP")

plt.show()
```

### Computing mean (absolute) difference over annotated brain structures

```python
reference_space = mcc.get_reference_space()
reference_space
```

```python
brain_summary_structures_id = 167587189
all_structures = reference_space.structure_tree.get_structures_by_set_id([brain_summary_structures_id])
all_structure_ids = [st['id'] for st in all_structures]
len(all_structure_ids)
```

```python
test_mask = reference_space.make_structure_mask([all_structure_ids[-1]])
plt.imshow(test_mask[100,:,:])
test_mask.sum()
```

```python
def calculate_mean_diff(diff_arr):
    mean_diff = np.zeros_like(annot).astype('float64')

    for id in all_structure_ids:
        mask = reference_space.make_structure_mask([id])
        diff_sum = (diff_arr*mask).sum()
        area_pixel_sum = mask.sum()
        area_mean_diff = diff_sum/area_pixel_sum
        mean_diff += area_mean_diff*mask
        print('Mean difference calculated for area',id)
    return mean_diff
```

```python
# VISpm_VISam_MAD = calculate_mean_diff(np.absolute(VISpm_pd-VISam_pd))
# np.save('VISpm_VISam_MAD.npy',VISpm_VISam_MAD)
VISpm_RSPagl_MAD = calculate_mean_diff(np.absolute(VISpm_pd-RSPagl_pd))
np.save('VISpm_RSPagl_MAD.npy',VISpm_RSPagl_MAD)
VISam_RSPagl_MAD = calculate_mean_diff(np.absolute(VISam_pd-RSPagl_pd))
np.save('VISam_RSPagl_MAD.npy',VISam_RSPagl_MAD)

VISpm_VISam_MD = calculate_mean_diff(VISpm_pd-VISam_pd)
np.save('VISpm_VISam_MD.npy',VISpm_VISam_MD)
VISpm_RSPagl_MD = calculate_mean_diff(VISpm_pd-RSPagl_pd)
np.save('VISpm_RSPagl_MD.npy',VISpm_RSPagl_MD)
VISam_RSPagl_MD = calculate_mean_diff(VISam_pd-RSPagl_pd)
np.save('VISam_RSPagl_MD.npy',VISam_RSPagl_MD)
```

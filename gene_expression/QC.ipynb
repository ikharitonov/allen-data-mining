```python
# last test: are there any experiments with empty structure unionizes when:
# - orientation_id == 2
# - failed == False
# are selected for all receptors
```

```python
def get_df(receptor):
    return pd.DataFrame(rma.model_query('SectionDataSet', criteria="[failed$eqfalse],products[abbreviation$eq'Mouse'],genes[acronym$eq'"+receptor+"'],probes[orientation_id$eq2]", include="probes(orientation),structure_unionizes"))
```

```python
data = {}
for r in tqdm(receptors):
    data[r] = get_df(r)
    data[r].to_csv(r+'.csv',index=False)
    if data[r].empty: print(r,'df empty')
    if data[r]['structure_unionizes'].empty: print(r,'structure unionizes df empty')
```

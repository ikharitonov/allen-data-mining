```python
import pandas as pd
from math import isnan
```

```python
# variable = 'expression_density'
variable = 'expression_energy'
```

```python
info_df = pd.read_csv('data/info.csv')
info_df
```

plane_of_section_id = 1 for coronal

plane_of_section_id = 2 for saggital

```python
sheet1_df = []
for receptor in info_df['receptor'].unique():
    exp_density_average = 0
    exps = 0
    for exp_id in info_df[(info_df['receptor']==receptor) & (info_df['has_data']==True)]['experiment_id']:
        exp_df = pd.read_csv('data/'+info_df[(info_df['receptor']==receptor) & (info_df['experiment_id']==exp_id)]['filename'].item())
        exp_density_average += exp_df[variable].item()
        exps+=1
    if exps==0:
        sheet1_df.append({'receptor': receptor,variable+'_average': 'N/A'})
        continue
    else: sheet1_df.append({'receptor': receptor,variable+'_average': exp_density_average / exps})
sheet1_df = pd.DataFrame(sheet1_df)
sheet1_df
```

```python
plane_dict = {1: 'coronal', 2: 'saggital'}

sheet2_df = []
for receptor in info_df['receptor'].unique():
#     exp_density_average = 0
#     exps = 0
    if info_df[info_df['receptor']==receptor]['experiment_id'].isnull().values.any():
        sheet2_df.append({'receptor': receptor,'experiment_id': 'N/A','plane': 'N/A',variable: 'N/A'})
        continue
    for exp_id in info_df[info_df['receptor']==receptor]['experiment_id']:
        exp_df = info_df[(info_df['receptor']==receptor) & (info_df['experiment_id']==exp_id)]
        if exp_df['has_data'].item() == False:
            sheet2_df.append({'receptor': receptor,'experiment_id': exp_id,'plane': plane_dict[int(exp_df['plane'].item())],variable: 'N/A'})
            continue
        union_data_df = pd.read_csv('data/'+exp_df['filename'].item())
        exp_density = union_data_df[variable].item()
        sheet2_df.append({'receptor': receptor,'experiment_id': exp_id,'plane': plane_dict[int(exp_df['plane'].item())],variable: exp_density})
#     if exps==0:
#         sheet2_df.append({'receptor': receptor,'expression_density_average': 'N/A'})
#         continue
#     else: sheet2_df.append({'receptor': receptor,'experiment_id': 0,'plane': 'coronal','expression_density': exp_density_average})
sheet2_df = pd.DataFrame(sheet2_df)
sheet2_df
```

```python
with pd.ExcelWriter(variable+'_data.xlsx') as writer:  
    sheet1_df.to_excel(writer, sheet_name='average_'+variable)
    sheet2_df.to_excel(writer, sheet_name='experiment_'+variable)
```

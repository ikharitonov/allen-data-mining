```python
from allensdk.core.mouse_connectivity_cache import MouseConnectivityCache
from allensdk.api.queries.ontologies_api import OntologiesApi
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
mcc = MouseConnectivityCache()
```

```python
structure_tree = mcc.get_structure_tree()
```

```python
oapi = OntologiesApi()
structure_set_ids = structure_tree.get_structure_sets()
pd.DataFrame(oapi.get_structure_sets(structure_set_ids))
```

## Getting list of experiements with injection in visual areas

```python
AREA_ID = 402

# injection_area = structure_tree.get_structures_by_id([AREA_ID])
experiments = mcc.get_experiments(injection_structure_ids=[AREA_ID])

print("%d total experiments found." % len(experiments))
```

```python
pd.DataFrame(experiments)
```

## Getting information (e.g., density, volume of projecting signal) with Structure Signal Unionization

```python
# https://alleninstitute.github.io/AllenSDK/allensdk.core.mouse_connectivity_cache.html#allensdk.core.mouse_connectivity_cache.MouseConnectivityCache.get_experiment_structure_unionizes
structure_unionizes = mcc.get_experiment_structure_unionizes(120437703, is_injection=None, structure_ids=[402], include_descendants=True)
print("%d structure unionizes." % len(structure_unionizes))
```

```python
pd.DataFrame(structure_unionizes)
```

Reference for unionization table columns: https://alleninstitute.github.io/AllenSDK/unionizes.html

### Selecting dense unionizes from this

```python
MIN_DENSITY = 0.5
MIN_VOLUME = 0

filtered_unionizes = structure_unionizes[structure_unionizes.projection_density > MIN_DENSITY]
filtered_unionizes = filtered_unionizes[filtered_unionizes.volume > MIN_VOLUME]
```

```python
large_structures = pd.DataFrame(structure_tree.nodes(filtered_unionizes.structure_id))
print("%d large and dense unionizes, %d structures" % ( len(large_unionizes), len(large_structures) ))
print(large_structures.name)
```

## Projection Matrix

```python
experiment_ids = [e['id'] for e in experiments]
children_areas = structure_tree.child_ids([AREA_ID])[0]

proj_matrix = mcc.get_projection_matrix(experiment_ids = experiment_ids, 
                               projection_structure_ids = children_areas,
                               parameter = 'projection_density')

row_labels = proj_matrix['rows'] # these are just experiment ids
column_labels = [ c['label'] for c in proj_matrix['columns'] ] 
matrix = proj_matrix['matrix']

fig, ax = plt.subplots(figsize=(15,15), facecolor='white')
heatmap = ax.pcolor(matrix, cmap=plt.cm.afmhot)

# put the major ticks at the middle of each cell
ax.set_xticks(np.arange(matrix.shape[1])+0.5, minor=False)
ax.set_yticks(np.arange(matrix.shape[0])+0.5, minor=False)

ax.set_xlim([0, matrix.shape[1]])
ax.set_ylim([0, matrix.shape[0]])          

# want a more natural, table-like display
ax.invert_yaxis()
# ax.xaxis.tick_top()

ax.set_xticklabels(column_labels, minor=False)
ax.set_yticklabels(row_labels, minor=False)

plt.xlabel('Areas')
plt.ylabel('Experiment IDs')
```
